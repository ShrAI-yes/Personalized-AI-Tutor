```python
import PyPDF2
import cv2
import pytesseract
import numpy as np
```

```python
pdf_file = open("ICSE-Class-6-Physics-Chapter-4-Pressure.pdf", "rb")
pdf_reader = PyPDF2.PdfReader(pdf_file)
```

```python
cv2.IMREAD_COLOR
```

```python
for page_num in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[page_num]
    xobj = page['/Resources']['/XObject'].getObject()

    for obj in xobj:
        if xobj[obj]['/Subtype'] == '/Image':
            data = xobj[obj].getData()
            image = cv2.imdecode(np.frombuffer(data, np.uint8), cv2.COLOR_BGR2GRAY)
            text = pytesseract.image_to_string(image)
            print(f"Text from page {page_num + 1}: {text}")
```
